```python
import numpy as np
import matplotlib.pyplot as plt

from DLP2000EVM_drivers.beagle_transfer import BeagleBoneImageClient
```

```python
client = BeagleBoneImageClient()
```

```python
H, W = client.height, client.width
n_masks = 180
rng = np.random.default_rng(12345)

A = (rng.random((H, W)) > 0.5).astype(np.uint8) * 255
B = (rng.random((H, W)) > 0.5).astype(np.uint8) * 255

### Alternate AAAABBBBAAAABBBB... (180 frames total)
masks = [(A if (i // 4 % 2 == 0) else B).copy() for i in range(n_masks)]

remote_blob = client.send_sequence_blob(masks, remote_file = "/home/debian/sequence.bin")
```

```python
client.play_sync_out(remote_file = remote_blob, frames = len(masks), gpio_in = 49, gpio_out = 60, loop = 1, warmup_edges = 2, timeout_ms = 10, pulse_each = 1)
```

```python
m = client.generate_mask("grid")
remote = client.send_image(m)
client.show_image(remote)
```
